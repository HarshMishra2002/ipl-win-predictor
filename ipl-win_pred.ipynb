```python
import numpy as np
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
match = pd.read_csv('matches.csv')
delivery = pd.read_csv('deliveries.csv')
```

```python
match.head()
```

```python
match.shape
```

```python
delivery.head()
```

```python
total_score_df = delivery.groupby(['match_id', 'inning']).sum()['total_runs'].reset_index()
```

```python
total_score_df.shape
```

```python
total_score_df.head()
```

```python
total_score_df = total_score_df[total_score_df['inning'] == 1]
```

```python
total_score_df
```

```python
match_df = match.merge(total_score_df[['match_id', 'total_runs']], left_on = 'id', right_on = 'match_id')
```

```python
match_df['team1'].unique()
```

```python
len(match_df['team1'].unique())
```

```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
match_df['team1'] = match_df['team1'].apply(lambda x: x.replace('Delhi Daredevils', 'Delhi Capitals') )
```

```python
match_df['team2'] = match_df['team2'].apply(lambda x: x.replace('Delhi Daredevils', 'Delhi Capitals') )
match_df['team1'] = match_df['team1'].apply(lambda x: x.replace('Deccan Chargers', 'Sunrisers Hyderabad') )
match_df['team2'] = match_df['team2'].apply(lambda x: x.replace('Deccan Chargers', 'Sunrisers Hyderabad') )
```

```python
len(match_df['team1'].unique())
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
len(match_df['team1'].unique())
```

```python
match_df
```

```python
match_df = match_df[match_df['dl_applied'] == 0]
```

```python
match_df = match_df[['match_id', 'city', 'winner', 'total_runs']]
```

```python
match_df
```

```python
delivery_df = match_df.merge(delivery, on = 'match_id')
```

```python
delivery_df
```

```python
delivery_df = delivery_df[delivery_df['inning'] == 2]
```

```python
delivery_df.shape
```

```python
delivery_df
```

```python
delivery_df['current_score'] = delivery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
delivery_df
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df
```

```python
delivery_df['runs_left'] = delivery_df['runs_left'].apply(lambda x: x+1)
```

```python
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
```

```python
delivery_df
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x: x if x == "0" else 1)
```

```python
delivery_df
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
wickets = delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
delivery_df['wickets'] = 10 - wickets
```

```python
delivery_df
```

```python
# current run rate  = runs/overs
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
# rrr
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
delivery_df
```

```python
delivery_df['result'] = delivery_df.apply(lambda x: 1 if x['winner'] == x['batting_team'] else 0, axis = 1)
```

```python
delivery_df
```

```python
delivery_df['result'].value_counts()
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
final_df
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df = final_df[final_df['balls_left'] != 0]
```

```python
final_df.sample()
```

```python
final_df.dropna(inplace = True)
```

```python
final_df
```

```python
X = final_df.iloc[:, : -1]
y = final_df.iloc[: , -1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state= 52)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps = [
    ('step1', trf),
    ('step2', LogisticRegression(solver = 'liblinear'))
])
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
